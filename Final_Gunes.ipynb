The operations here will be done in python and using the jupyter notebook. For these operations, we first import the following modules so that we can create, map and visualize the matrices:  

```python
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

### 1.Creating 100x100 numpy array (100x100 matrix)

To generate a numpy array with 100 rows, each row containing 100 random integers from 0 to 100, *random* and *randint* commands are used and displayed with *print*.  

```python
a = np.random.randint(0, 100, size=(100, 100))
# a is a variable (object)

print(a)
print(a.shape)
```

as *output*, obtained 100x100 matrix with random integers from 0 to 100.

### 2.Making a heatmap of 100x100 array

We can convert our 100x100 matrix, which we call *'a'*, into a heatmap with *heatmap* command. 

```python
ax = sns.heatmap(a, linewidth=0.00001)

plt.show()
```

Obtained heatmap is displayed with the *plt.show* command and all numbers are included in the heat map from the darkest at 0 to the lightest at 100. 

### 3.Filtering the odd numbers

Using the same matrix, the function *'a % 2 == 1'* is used to filter out odd numbers in the matrix. The matrix in which odd numbers are filtered is named *'filter_a'*. In addition, the place of these odd numbers in the array was preserved with the * function and created as *'new_a'*. 

```python
filter_a = a % 2 == 1

new_a = a*filter_a

print(new_a)
print(filter_a)
print(filter_a.shape)
```

With the *print* command, a filtered matrix in which odd numbers are shown as 'true' is obtained as output. It is seen that our 100x100 matrix is preserved with the shape command. The *print(new_a)* command outputs an array that shows only odd numbers in their original places. 

### 4.Converting True/False statements into [0,1] values

Filtered array (filter_a) is converted to [0, 1] with *astype(int)* command and odd numbers are assigned as *1* 

```python
filter_a.astype(int)
```

With the *shape* function we can check that our 100x100 matrix is preserved 

```python
print(filter_a.astype(int).shape)
```

### 5.Making a heatmap of new 100x100 array

Heatmap is made from the array converted to [0, 1] values with the *heatmap* command. 

```python
ax = sns.heatmap(filter_a.astype(int), linewidth=0.00001)

plt.show()
```

Obtained heatmap is displayed with the *plt.show* command and *0 and 1* two numbers represented by the darkest and lightest colors. We get the same heatmap in the filtered array (represented as true and false) as seen below:  

```python
ax = sns.heatmap(filter_a, linewidth=0.00001)

plt.show()
```

Finally, as seen below; If we wanted to filter out even numbers, we would have to use the formula *'a % 2 == 0'* this time. We would have the inverse of our matrix and a complementary heatmap. Thus, we can also verify our transaction. 

```python
filter_a2 = a % 2 == 0
new_a2 = a*filter_a2
print(filter_a2)
print(new_a2)
```
